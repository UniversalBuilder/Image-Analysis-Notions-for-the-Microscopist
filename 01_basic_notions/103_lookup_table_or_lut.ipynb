## 103/ Lookup table or LUT : viewing pixel values in color.

If you consider the simple case of an 8-bit grayscale digital image, a single number per pixel represents the intensity of the signal. 

> *Without additional information, **there is no specific color tied to the pixel values in the data itself**.*

You can represent these values using gray levels, which is often the default in life science. But **you can also use a LUT (lookup table) or Colormap** to assign a specific color to each value and create an **artificial color** representation of the data.

> *Some colormaps are **gradients** from dark (low intensities) to light (high intensities), but there are also specialized colormaps than can **highlight** a change in intensity or texture, **split your data into several classes**, or **mimic reality** with a colorized version of a black and white image like a satellite image of an island.*

**Here is the exact same image, but represented using different colormaps:**

- gray : *the default color map*
- jet : *highlights high intensities*
- Set1 : *divides image into classes*
- flag : *highlights areas with high or low variabilities (texture)*
- terrain : *mimics a satellite view of land*

```python
import numpy as np
import pandas as pd
from skimage import io
import matplotlib.pyplot as plt

big_image = io.imread('images/first-instar-brain.tif') 


fig, ax = plt.subplots(figsize=(25, 4), ncols=5)

img = big_image[20,:,:,1]

map1 = ax[0].imshow(img.astype(float), cmap='gray')
ax[0].set_title('gray')
plt.colorbar(map1, ax=ax[0])

map2 = ax[1].imshow(img.astype(float),cmap='jet')
ax[1].set_title('jet')
fig.colorbar(map2,ax=ax[1])

map3 = ax[2].imshow(img.astype(float),cmap='Set1')
ax[2].set_title('Set1')
fig.colorbar(map3, ax=ax[2])

map4 = ax[3].imshow(img.astype(float),cmap='flag')
ax[3].set_title('flag')
fig.colorbar(map4, ax=ax[3])

map5 = ax[4].imshow(img.astype(float),cmap='terrain')
ax[4].set_title('terrain')
fig.colorbar(map5, ax=ax[4]);
```
